```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.meilleursagents.com/prix-immobilier/ile-de-france/"
page=requests.get(url)
print(page.status_code==200)
# page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
# print(soup.prettify)
```

```python
tables = soup.find_all('table')
# tables
```

```python
dfs = pd.read_html(str(tables))
```

```python
# VILLE Prix/Loyer
prix_ville = dfs[0]
# DEPARTEMENT Prix/Loyer
prix_dep = dfs[1]
# REGION Prix/Loyer
prix_region = dfs[2]
```
